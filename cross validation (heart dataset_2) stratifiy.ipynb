```python
import numpy as np
import pandas as pd
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
l=LabelEncoder()
dataset["Sex"]=l.fit_transform(dataset["Sex"])
dataset["Sex"].value_counts()
```

```python
dataset["ChestPainType"]=l.fit_transform(dataset["ChestPainType"])
dataset["ChestPainType"].value_counts()
```

```python
dataset["RestingECG"]=l.fit_transform(dataset["RestingECG"])
dataset["RestingECG"].value_counts()
```

```python
dataset["ExerciseAngina"]=l.fit_transform(dataset["ExerciseAngina"])
dataset["ExerciseAngina"].value_counts()
```

```python
dataset["ST_Slope"]=l.fit_transform(dataset["ST_Slope"])
dataset["ST_Slope"].value_counts()
```

```python
data=dataset
data
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
from sklearn.model_selection import StratifiedKFold
sf=StratifiedKFold(n_splits=4,random_state=10,shuffle=True)
sf.get_n_splits(x,y)
print(sf)
```

```python
for train_index,test_index in sf.split(x,y):
    print("TRAIN:", train_index, "TEST:", test_index)
    x_train, x_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
from sklearn.linear_model import LogisticRegression
l1=LogisticRegression()
from sklearn.model_selection import cross_val_score
scores=cross_val_score(l1,x_train,y_train,cv=sf)
print(scores)
print(np.mean(scores)*100)
```

```python
from sklearn.model_selection import cross_val_predict
y_pred=cross_val_predict(l1,x_test,y_test,cv=sf)
print(y_pred)
```
