```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\Sushant\Desktop\Ankita\Datasets\Mall_Customers.csv")
data
```

```python
x=data.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import KMeans
ab=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=10)
    kmeans.fit(x)
    ab.append(kmeans.inertia_)
    print(ab)
plt.plot(range(1,11),ab)    
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,init="k-means++",random_state=10)
ymeans=kmeans.fit_predict(x)
ymeans
```
